Columna **Cruce** es una variable categorica, de tipo string, que contiene 317 datos distintos y ademas un 24,6% de valores nulos. Esta variable contien los datos de cuando un hecho ocurre en un cruce, debido a esto se decide realizar una columna binaria, con un Si o No, dependiendo si el hecho ocurrio en un cruce o no y ademas mantener la columna actual para su posterios analisis y rellenar los NAN con 'No en cruce'.

```python
# Se ven los valores nulos de cruce en relación con 'Lugar del hecho' y 'Tipo de calle'
homicidios_hechos[homicidios_hechos['Cruce'].isnull()][['Cruce', 'Lugar_Del_Hecho', 'Tipo_De_Calle']][:5]
```

```python
# Se ven los valores NO nulos de cruce en relación con 'Lugar del hecho' y 'Tipo de calle'
homicidios_hechos[homicidios_hechos['Cruce'].notnull()][['Cruce', 'Lugar_Del_Hecho', 'Tipo_De_Calle']][:5]
```

```python
# Rellenamos los NaN con el valor 'No en cruce'
homicidios_hechos['Cruce'].fillna('No en cruce', inplace=True)

# Cambiamos el nombre de la columna 'Cruce' por 'Nombre_Cruce'
homicidios_hechos.rename(columns={'Cruce': 'Nombre_Cruce'}, inplace=True)

# Creamos la columna 'Cruce' binaria donde contendrá valores 'Sí' o 'No'
homicidios_hechos['Cruce'] = 'Sí'  # Establecemos inicialmente todos los valores como 'Sí'
homicidios_hechos.loc[homicidios_hechos['Nombre_Cruce'] == 'No en cruce', 'Cruce'] = 'No'
```

```python
# Se revia la columna
homicidios_hechos[['Cruce','Nombre_Cruce', 'Lugar_Del_Hecho', 'Tipo_De_Calle']]
```

La tercer columna con valores faltantes es 'Dirección_Normalizada' tiene solo 8 datos nulos, lo que representa el 1.1% de los datos de la columna. Se orservan esos registros a continuación.

```python
homicidios_hechos[homicidios_hechos['Dirección Normalizada'].isnull()][['Dirección Normalizada', 'Lugar_Del_Hecho', 'Tipo_De_Calle']][:10]
```

No tenemos informacion que nos permita determinar un dato concreto de estos faltantes, pero no es conveniente eliminas los registros. Por lo tanto, se procede a rellenar estos valores con SD, indicando que es Sin Dato

```python
# Se rellenan nulos con SD
homicidios_hechos['Dirección Normalizada'].fillna("SD", inplace=True)
# Se verifican los datos imputados
homicidios_hechos[homicidios_hechos['Dirección Normalizada']== 'SD'][['Dirección Normalizada', 'Lugar_Del_Hecho', 'Tipo_De_Calle']]
```

como podemos observar 'Hora', 'Hh' y 'Calle' contienen varios tipos de datos en la misma columna, por lo que procedemos a revisarlas y corregirlas. 

Columna 'Hora'

```python

# Cantidad de valores por tipo de dato en la columna 'Hora'
conteo_tipos = homicidios_hechos['Hora'].apply(type).value_counts()

#muestra de cada tipo de dato
for tipo, cantidad in conteo_tipos.items():
    muestra = homicidios_hechos.loc[homicidios_hechos['Hora'].apply(type) == tipo, 'Hora'].head()
    print(f'Tipo: {tipo}, Cantidad: {cantidad}\nMuestra:\n{muestra}\n')
```

Dado que la mayoría de los datos son del tipo datetime.time, se decide convertir todos a este tipo de dato.

```python
# Se cambia el tipo de dato
homicidios_hechos['Hora'] = homicidios_hechos['Hora'].apply(lambda x: convertir_a_time(x))
# Se verifica la cantidad de valores por tipo de dato en la columna 'hora'
print('Tipos de datos:')
print(homicidios_hechos['Hora'].apply(type).value_counts())
print('Registro con NoneType:')
homicidios_hechos[homicidios_hechos['Hora'].isna()]
```

se observa que no se encuentras datos en Hora y Hh. Para rellenar este dato con la moda de la columna Hora

```python
# calcula la hora mas común
hora_moda = homicidios_hechos['Hora'].mode().iloc[0]
print(f'La hora mas común es: {hora_moda}')
# Se reemplaza el valor None por la hora más común
homicidios_hechos['Hora'].fillna(hora_moda, inplace=True)
# se verifica el tipo de dato para la columna
homicidios_hechos['Hora'].apply(type).value_counts()
```
